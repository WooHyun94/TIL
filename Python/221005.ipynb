```python
x2 = [1, 2, 3]
y2 = [1, 50, 200]
```

```python
fig, axe1 = plt.subplots()

axe2 = axe1.twinx()
chart1 = axe1.plot(x1, y1, color = 'red', label = 'data1')
chart2 = axe2.plot(x2, y2, color = 'blue', label = 'data2')

axe1.set_xlabel('x', fontsize = 15)
axe1.set_ylabel('y1', fontsize = 15)
axe2.set_ylabel('y2', fontsize = 15)

chart = chart1 + chart2
axe1.legend(chart, ['data1', 'data2'])
```

```python
plt.figure(figsize = (8, 6))
sns.barplot(data = df, x = '', y = '', palette = 'Set2')
plt.show()
```

```python
plt.figure(figsize = (8, 6))
sns.barplot(data = df, x = '', y = '', order = df.sort_values('', ascending = False).)
plt.show()
```

```python
plt.figure(figsize = (8, 6))
chart = sns.barplot(data = df, x = '', y = '', order = df.sort_values('', ascending = False).)
ylabels = ['{:.0f}'.format(i) + ' ' for i in chart.get_yticks() / 10000]
chart.set_yticklabels(ylabels)
plt.show()
```

```python
plt.figure(figsize = (8, 6))
chart = sns.barplot(data = df, y = '', x = '', order = df.sort_values('', ascending = False).)
ylabels = ['{:.0f}'.format(i) + ' ' for i in chart.get_yticks() / 10000]
chart.set_yticklabels(ylabels)
plt.show()
```

```python
plt.figure(figsize = (8, 6))
chart = sns.barplot(data = df, y = '', x = '', order = df.sort_values('', ascending = False)., color = 'blue')
ylabels = ['{:.0f}'.format(i) + ' ' for i in chart.get_yticks() / 10000]
chart.set_yticklabels(ylabels)
plt.show()
```

```python
plt.figure(figsize = (8, 6))
chart = sns.barplot(data = df, y = '', x = '', order = df.sort_values('', ascending = False)., color = 'blue')
xlabels = ['{:.0f}'.format(i) + ' ' for i in chart.get_xticks() / 10000]
chart.set_xticklabels(xlabels)

plt.xlabel("", fontsize = 15)
plt.ylabel('', fontsize = 15)
plt.title("  ", fontsize = 20)
plt.show()
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length')
plt.grid(True, alpha = 0.7)
```

```python
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species')
plt.grid(True, alpha = 0.5)
```

```python
def function1(x): 
    if x < 3:
        return 's'
    elif x < 5:
        return 'm'  
    else:
        return 'l' 
```

```python
df['petal_length2'] = df['petal_length'].apply(function1)
```

```python
df
```

```python
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species', style = 'petal_length2')
plt.grid(True, alpha = 0.5)
```

```python
def function2(x): 
    if x < 3:
        return 3
    elif x < 5:
        return 5  
    else:
        return 7
```

```python
df['petal_length3'] = df['petal_length'].apply(function2)
```

```python
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species', size = 'petal_length3')
plt.grid(True, alpha = 0.5)
```

```python
sns.relplot(data = df, x = 'sepal_width', y = 'sepal_length', col = 'species')
```

```python
sns.lmplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species')
plt.grid(True, alpha = 0.5)
```
