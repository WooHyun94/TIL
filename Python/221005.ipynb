```python
# subplot
plt.figure(figsize = (6, 8))
plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.title("data1")

plt.subplot(2, 1, 2)
plt.plot(x2, y2)
plt.title("data2")
plt.tight_layout()
```

```python
x2 = [1, 2, 3]
y2 = [1, 50, 200]
```

```python
fig, axe1 = plt.subplots()

axe2 = axe1.twinx()
chart1 = axe1.plot(x1, y1, color = 'red', label = 'data1')
chart2 = axe2.plot(x2, y2, color = 'blue', label = 'data2')

axe1.set_xlabel('x', fontsize = 15)
axe1.set_ylabel('y1', fontsize = 15)
axe2.set_ylabel('y2', fontsize = 15)

chart = chart1 + chart2
axe1.legend(chart, ['data1', 'data2'])
```

```python
movie_title = ['크루엘라', '극장판 귀멸의 칼날 : 무한열차편', '학교 가는 길']
audience = [664308, 2099131, 20067]

data = {'영화제목' : movie_title, '누적관객' : audience}
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize = (8, 6))
sns.barplot(data = df, x = '영화제목', y = '누적관객', palette = 'Set2')
plt.show()
```
