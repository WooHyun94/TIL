```python
import pandas as pd  
import numpy as np   
import copy
```

```python
index = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame({'a' : [i for i in range(1, 11)]
                , 'b' : [i for i in range(11, 21)]
                , 'c' : [i for i in range(21, 31)]}, index = index)
df
```

```python
# a가 3 이하 이거나 7이상인 데이터를 출력하시오
df[(df['a'] <= 3) | (df['a'] >= 7)]
```

```python
# a가 3 이상이고, b가 16미만이거나 c가 30인 데이터를 출력하시오.

df[(df['a'] >= 3) & ((df['b'] < 16) | (df['c'] == 30))]
```

```python
df = pd.DataFrame({'a' : [2, 3, 2, 7, 4],
                    'b' : [2, 1, 3, 5, 3], 
                    'c' : [1, 1, 2, 3, 5]})
df
```

```python
df = pd.DataFrame({'a' : [2, 3, 2, 7, 4],
                    'b' : [2, 1, 3, 5, 3], 
                    'c' : [1, 1, 2, 3, 5]})
df
```

```python
# 값 기준 정렬
# a열 기준으로 오름차순 정렬
df.sort_values(by = ['a'], inplace = True)
```

```python
df
```

```python
# a열 기준으로 내림차순 정렬
df.sort_values(by = ['a'], ascending = False, inplace = True)
```

```python
df
```

```python
df.sort_values(by = ['a', 'b'])
```

```python
df.sort_values(by = ['a', 'b'], ascending = [True, False])
```
