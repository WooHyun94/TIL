- 서버리스 
    - 서버를 고려하지 않고 애플리케이션과 서비스를 구축 가능
    - 유연한 확장
    - 종량제 요금
    - 자동화된 고가용성

- 소프트 스킬
    - 개인적인 특징과 성격처럼 정량화 할 수 없는 특성을 말한다.
    - 행도, 태도, 가치에 뿌리를 두고 있다

- 시간 관리의 중요성
    - 컨퍼런서 등을 꾸준하게 시청 및 블로그 깃허브에 정리하는 습관이 필요

- 좋은 습관을 만들기 위해 이벤트 뒤에 원하느 행동을 넣음으로써 습관화하기

- Q : 백엔드 엔지니어가 처리하는 클라우드 업무 & 클라우드 엔지니어가 처리하는 클라우드 업무 
    - A : 백엔드 엔지니어는 어플리케이션의 안정적인 구동에 클라우드 엔지니어는 서버의 안정적 구동에 힘을 쏟는다
- Q : 백엔드의 트랜드를 따라잡기 위한 공부 방법
    - A : 채용 공고를 중점적으로 파악하라. 채용 공고만큼 트랜드가 잘 반영된 곳이 없다
    - A : 유튜브의 글로벌 컨퍼런스를 시청하라 
- Q : 필요한 핵심 역량
    - A : 어플리케이션에 더욱 가까이 있기에 비즈니스 로직과 메모리 동작 구조 등 안정적인 서비스 유지에 대한 역량을 키워라 

### 클라우드 엔지니어 로드맵

- 클라우드 엔진어가 하는 일 
    - 클라우드 인프라 환경을 구축하고 운영
    - 기준 인프라 환경이 있다면 이를 클라우드 환경으로 변환
    - 지속적으로 클라우드를 운영관리하고 유지보수 및 문제를 해결
    - 자동화 기룻을 통해 적합한 인프라 서비스 환경을 꾸준히 개선

- AWS의 클라우드 엔지니어 유형
    - Solutions Architect
    - Clud Architect
    - Account Manager
    - Cloud Support Engineer

- 어떻게 하면 좋은 개발자가 될까?
    - 사이드 프로젝트를 통한 성장 
        - 프로젝트 기술스택 잡기
            - 채용 공고 
        - 팀 구성
        - 주제 선정
        - 프로젝트 진행
        - 진행한 프로젝트 정리 

- 채용 공고를 분석하는 것이 중요

- 클라우드 위에 올라가는 서비스에 대한 이해
