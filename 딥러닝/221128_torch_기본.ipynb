```python
import torch
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
```

```python
train_data = datasets.MNIST(
    root='data',
    train=True,
    download=True,
    transform=transforms.ToTensor()
)

test_data=datasets.MNIST(
    root='data',
    train=False,
    download=True,
    transform=transforms.ToTensor()
)
```

```python
train_loader = DataLoader(train_data, batch_size=128, shuffle=True)
test_loader = DataLoader(test_data, batch_size=128, shuffle=False)
```

```python
data_iteration = iter(train_loader)
images, labels = data_iteration.next()

print(images.shape)
print(labels.shape)
```

```python
plt.imshow(images[0].reshape(28, 28))
print(images[0].shape)
print(labels[0])
```

```python
from torch.utils.data import Dataset, DataLoader
import torch
import pandas as pd
```

```python
from random import sample
class My_Dataset(Dataset):
    def __init__(self, data, labels):
        self.data = data
        self.labels = labels
    def __len__(self):
        return len(self.labels)
    def __getitem__(self, idx):
        data = self.data[idx]
        label = self.labels[idx]
        sample = {'Data' : data, "Class" : label}
        return sample
```

```python
data = ['New York', 'Boston', 'California', 'Boston', 'New York']
labels = [1, 2, 3, 2, 1]

Data_frame = pd.DataFrame({'Data':data, "Labels": labels})

Data = My_Dataset(Data_frame['Data'], Data_frame['Labels'])
```
