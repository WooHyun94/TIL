```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('./data/image.jpeg')
image
```

```python
plt.imshow(image)
```

```python
image_cvt = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_cvt)
```

```python
image_gray = cv2.cvtColor(image_cvt, cv2.COLOR_RGB2GRAY)

plt.imshow(image_gray, cmap='gray')
```

```python
plt.figure(figsize=(15,8))

plt.subplot(1, 3, 1)
plt.imshow(image)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image_cvt)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(image_gray, cmap='gray')
plt.axis('off')
```
