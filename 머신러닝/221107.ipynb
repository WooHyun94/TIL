## 클러스터링(Clustering)
- 비슷한 데이터끼리 묶는 방법
- 클러스터링의 엄밀한 정의 
  - Cluster analysis or clustering is the task of grouping a set of objects in such a way that objects in the same group are more similar to each other than to those in other groups

1. data representation (feature vector)
2. similarity measure (distance function)

### 1) K-Means Clustering
- 가장 직관적이고 이해하기 쉬운 클러스터링 방법
- K-평균 알고리즘 이라고도 부르며, 평균을 기준으로 데이터들을 묶어줌
- 엄청나게 빠른 속도로 수행되며, 결과가 무조건 나온다는 것이 보장되어 있음

## 피처 엔지니어링 (feature engineering)
- 피처 엔지니어링은 데이터를 분석하고자 하는 목적에 맞게 가공하는 과정이다.
- 분석가는 데이터에서 분석에 필요한 피처를 추출하여 모델링에 사용한다.
- 어떤 피처를 추출하느냐는 모델링을 통한 피드백과 분석가의 직관에 의해 이루어진다.
- 피처를 잘못 선택했을 때 발생하는 문제점들에 대해 인지한다.

### 차원의 저주 
- 고차원 데이터에 취약한 머신러닝 문제
  - feature space의 차원이 커질수록 머신러닝 성능이 저하되는 문제
  - 과적합될 가능성이 높아지며, 유의미한 패턴을 찾기가 힘들어짐
  - 차원 감소 기법을 이용하여 저차원 공간으로 변환
  - PCA(Principal Component Analysis), AutoEncoder 등을 많이 사용
